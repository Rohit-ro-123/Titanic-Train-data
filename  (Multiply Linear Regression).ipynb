```python
X=data_1[["Hiring","Patents"]]
y=data_1[["Net Income"]]
```

```python
import statsmodels.api as sm
```

```python
X_1=sm.add_constant(X)
```

```python
X_1.head()
```

```python
lm=sm.OLS(y,X_1)
lm=lm.fit()
```

```python
lm.summary()
```

```python
lm.params
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=21)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
X_test_1=sm.add_constant(X_test)
y_pred=lm.predict(X_test_1)
y_pred
```

```python
X_train_1=sm.add_constant(X_train)
lm=sm.OLS(y_train,X_train_1)
lm=lm.fit()
```

```python
lm.summary()
```

```python
X="Hiring"
plt.figure(figsize=(15,5))
plt.scatter(X_train["Hiring"],lm.predict(X_train_1),color="red")
plt.scatter(X_train["Hiring"],y_train,color="blue")
plt.scatter(X_test["Hiring"],y_test,color="grey")
plt.scatter(X_test["Hiring"],y_test,color="black")
plt.grid()
plt.show()
```

```python
X="Patents"
plt.figure(figsize=(15,5))
plt.scatter(X_train["Patents"],lm.predict(X_train_1),color="red")
plt.scatter(X_train["Patents"],y_train,color="blue")
plt.scatter(X_test["Patents"],y_test,color="grey")
plt.scatter(X_test["Patents"],y_test,color="black")
plt.grid()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
testing_data=pd.DataFrame()
testing_data
```

```python

testing_data=X_test.copy()
testing_data["Actuals"]=y_test["Net Income"]
testing_data["Predictions"]=y_pred
testing_data["Error"]=testing_data["Actuals"]-testing_data["Predictions"]
testing_data
```

```python
mse=mean_squared_error(testing_data["Actuals"],testing_data["Predictions"])
print("MSE:",round(mse,2))
```

```python
rsme=np.sqrt(mse)
print("RSME:",round(rsme,2))
```

```python
from sklearn.metrics import r2_score
```

```python
rsq=r2_score(testing_data["Actuals"],testing_data["Predictions"])
print("R square:",round(rsq,3))
```
