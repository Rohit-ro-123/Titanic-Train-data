# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
X=data_1[["Age","Fare"]]
y=data_1[["Survived"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , train_size = 0.8, random_state = 100)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
dt=DecisionTreeClassifier()
dt=dt.fit(X_train,y_train)
```

```python
y_train_prob=dt.predict_proba(X_train)[:,1]
y_train_prob
```

```python
y_train_pred=dt.predict(X_train)
y_train_pred
```

```python
X_train_op=X_train.copy()
X_train_op["Y_train_actuals"]=y_train["Survived"]
X_train_op["Y_train_pred"]=y_train_pred
X_train_op["Y_train_prob"]=y_train_prob
X_train_op.tail()
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(X_train_op["Y_train_actuals"],X_train_op["Y_train_pred"])
```

```python
from sklearn.metrics import precision_score
```

```python
#precision_score(X_train_op["Y_train_actuals"],X_train_op["Y_train_pred"])
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cnf_matrix = confusion_matrix(X_train_op["Y_train_actuals"], X_train_op["Y_train_pred"])

sns.heatmap(cnf_matrix, annot = True, cmap = "YlGnBu", fmt='.3g')
plt.ylabel("Actuals", fontsize = 13)
plt.xlabel("Predictions", fontsize = 13)
plt.title("Confusion Matrix", fontsize = 14)
plt.show()
```

# Random Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=data_1[["Age","Fare"]]
y=data_1[["Survived"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , train_size = 0.8, random_state = 100)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
rf=RandomForestClassifier(max_depth=2,random_state=12345)
rf=rf.fit(X_train,y_train)
```

```python
y_train_prob=rf.predict_proba(X_train)[:,1]
y_train_prob
```

```python
y_train_pred=rf.predict(X_train)
y_train_pred
```

```python
X_train_op=X_train.copy()
X_train_op["Y_train_actuals"]=y_train["Survived"]
X_train_op["Y_train_pred"]=y_train_pred
X_train_op["Y_train_prob"]=y_train_prob
X_train_op.tail()
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(X_train_op["Y_train_actuals"],X_train_op["Y_train_pred"])
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cnf_matrix = confusion_matrix(X_train_op["Y_train_actuals"], X_train_op["Y_train_pred"])

sns.heatmap(cnf_matrix, annot = True, cmap = "YlGnBu", fmt='.3g')
plt.ylabel("Actuals", fontsize = 13)
plt.xlabel("Predictions", fontsize = 13)
plt.title("Confusion Matrix", fontsize = 14)
plt.show()
```

```python
rf.n_estimators
```

```python
rf.estimators_[0:2]
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(10,5),dpi=900)
for index in range(0,2):
    plot_tree(rf.estimators_[index],filled=True,fontsize=5,ax=axes[index],feature_names=X_train.columns.tolist(),class_names=["0","1"])
plt.show()
```
